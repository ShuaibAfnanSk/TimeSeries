```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistics:{}'.format(result[0]))
    print('P-value:{}'.format(result[1]))
    if result[1] <= 0.05:
        print("Reject the Null Hypothesis")
    else:
        print("Accept the Null Hypothesis")
```

```python
adf_test(df['volume'])
```

```python
df['difference of Volume'] = df['volume'] - df['volume'].shift(1)
```

```python
df = pd.read_csv('iwm_dataset.csv')
```

```python
df.head()
```

```python
future_datest_df=pd.read_csv('future_dates.csv')
future_datest_df.head()
```

```python
dataset = pd.concat([df,future_datest_df])
```

```python
dataset = dataset.drop(columns=['idx'],axis=1)
```

```python
dataset.tail()
```

```python
dataset = dataset.set_index('date')
```

```python
dataset.index = pd.to_datetime(dataset.index)
```

```python
sarimax_model = sm.tsa.statespace.SARIMAX(dataset['volume'],order=(5,2,1),seasonal_order=(5,2,1,12))
results = sarimax_model.fit()
dataset['forecast'] = results.predict(start=366,end=454,dynamic=True)
dataset[['volume','forecast']].plot()
```

```python
sarimax_model = sm.tsa.statespace.SARIMAX(dataset['level'],order=(0,1,0),seasonal_order=(0,1,0,12))
results = sarimax_model.fit()
dataset['forecast'] = results.predict(start=366,end=454,dynamic=True)
dataset[['level','forecast']].plot()
```

```python
sarimax_model = sm.tsa.statespace.SARIMAX(dataset['usage'],order=(5,2,1),seasonal_order=(5,2,1,12))
results = sarimax_model.fit()
dataset['forecast'] = results.predict(start=366,end=454,dynamic=True)
dataset[['usage','forecast']].plot()
```
